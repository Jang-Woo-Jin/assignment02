# import packages for plotting graphs and manipulating data:

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

# define my function: f(x)=-xcos(x)+sin(x)

```python
def myFunction(x):
    f = -x*np.cos(x)+np.sin(x)
    return f
```

# define the derivative of my function: f'(x)=xsin(x)

```python
def myDerivativeFunction(x):
    Df = np.sin(x)*x
    return Df
```

# define the domain of the function:x=[âˆ’10:0.1:10]

```python
x = np.arange(-10, 10, 0.1)
```

# compute the graph

```python
f = myFunction(x)
Df = myDerivativeFunction(x)
```

# pick 3 points in the domain

```python
point1=math.pi
point2=math.pi/2
point3=-math.pi*1.9
```

# calculate Taylor approximation of 3 points

```python
point1_ta=myFunction(point1)+myDerivativeFunction(point1)*(x-point1)
point2_ta=myFunction(point2)+myDerivativeFunction(point2)*(x-point2)
point3_ta=myFunction(point3)+myDerivativeFunction(point3)*(x-point3)
```

# plot the graphs for the function and its derivative

```python
plt.plot(x, f, 'black', label="function")
plt.plot(x, Df, 'grey', label="derivative")
plt.plot(x, point1_ta,'r', label="point1_taylor approximation")
plt.plot(x, point2_ta,'orange', label="point2_taylor approximation")
plt.plot(x, point3_ta,'y', label="point3_taylor approximation")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```
